#### 3.2 hdf5的读取存储
支持压缩（blosc）
```
DataFrame.to_hdf(path_or_buf, key, mode='a', complevel=None, complib=None, append=False, format=None, index=True, min_itemsize=None, nan_rep=None, dropna=None, data_columns=None, errors='strict', encoding='UTF-8')
```
1. keystr
    Identifier for the group in the store.
1. mode{‘a’, ‘w’, ‘r+’}, default ‘a’
     Mode to open file:

    1. ‘w’: write, a new file is created (an existing file with the same name would be deleted).
    1. ‘a’: append, an existing file is opened for reading and writing, and if the file does not exist it is created.
    1. ‘r+’: similar to ‘a’, but the file must already exist.
1. complib{‘zlib’, ‘lzo’, ‘bzip2’, ‘blosc’}, default ‘zlib’
    Specifies the compression library to be used. As of v0.20.2 these additional compressors for Blosc are supported (default if no compressor specified: ‘blosc:blosclz’): {‘blosc:blosclz’, ‘blosc:lz4’, ‘blosc:lz4hc’, ‘blosc:snappy’, ‘blosc:zlib’, ‘blosc:zstd’}. #### 3.3 json的读取存储

```
DataFrame.to_json(path_or_buf=None, orient=None, date_format=None, double_precision=10, force_ascii=True, date_unit='ms', default_handler=None, lines=False, compression='infer', index=True, indent=None, storage_options=None)
```
1. orientstr:Indication of expected JSON string format. 
    1. Series:
        default is ‘index’
        allowed values are: {‘split’, ‘records’, ‘index’, ‘table’}.
    1. DataFrame:
        default is ‘columns’
        allowed values are: {‘split’, ‘records’, ‘index’, ‘columns’, ‘values’, ‘table’}.

    1. The format of the JSON string:
        1. ‘split’ : dict like {‘index’ -> [index], ‘columns’ -> [columns], ‘data’ -> [values]}

        1. ‘records’ : list like [{column -> value}, … , {column -> value}]

        1. ‘index’ : dict like {index -> {column -> value}}

        1. ‘columns’ : dict like {column -> {index -> value}}

        1. ‘values’ : just the values array

        1. ‘table’ : dict like {‘schema’: {schema}, ‘data’: {data}}

    Describing the data, where data component is like orient='records'.
1. date_format{None, ‘epoch’, ‘iso’}
     Type of date conversion. ‘epoch’ = epoch milliseconds, ‘iso’ = ISO8601. The default depends on the orient. For orient='table', the default is ‘iso’. For all other orients, the default is ‘epoch’.

````python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                   'key1':['K0','K0','K1','K2'],
                   'key2':['K0','K1','K0','K1']})
right = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3'],
                   'key1':['K0','K1','K1','K2'],
                   'key2':['K0','K0','K0','K0']})
left
````

````python
right
````

````python
pd.merge(left,right,on=['key1','key2'])#on 按什么进行拼接 how 怎么拼接:inner内连接default，
#保留left["key1"]==right["key1"]&&left["key2"]==right["key2"]
````

````python
pd.merge(left,right,on=['key1','key2'],how="outer")#on 按什么进行拼接 how 怎么拼接:outer，所有数据都保留
````

````python
pd.merge(left,right,on=['key1','key2'],how="left")#on 按什么进行拼接 how 怎么拼接:left，
#outerh后保留left!=nan
````

````python
pd.merge(left,right,on=['key1','key2'],how="right")#on 按什么进行拼接 how 怎么拼接:right，
#outerh后保留right!=nan
````
